```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definir variveis fuzzy (entrada e sada)
entrada = ctrl.Antecedent(np.arange(0, 4, 0.01), 'entrada')
entrada['N'] = fuzz.gaussmf(entrada.universe, 1.5, 0.16)
entrada['M'] = fuzz.gaussmf(entrada.universe, 2.0, 0.16)
entrada['F'] = fuzz.gaussmf(entrada.universe, 2.5, 0.16)

saida = ctrl.Consequent(np.arange(0, 4, 0.01), 'saida')
saida['N'] = fuzz.gaussmf(saida.universe, 1.5, 0.16)
saida['M'] = fuzz.gaussmf(saida.universe, 2.0, 0.16)
saida['F'] = fuzz.gaussmf(saida.universe, 2.5, 0.16)

# Definir regras fuzzy
regra1 = ctrl.Rule(entrada['N'], saida['N'])
regra2 = ctrl.Rule(entrada['M'], saida['M'])
regra3 = ctrl.Rule(entrada['F'], saida['F'])

# Criar sistema de controle
sistema = ctrl.ControlSystem([regra1, regra2, regra3])
simulacao = ctrl.ControlSystemSimulation(sistema)

# Teste
simulacao.input['entrada'] = 0.1
simulacao.compute()
print(simulacao.output['saida'])
```
